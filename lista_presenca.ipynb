```python
import pandas as pd
```

```python
#df_league_2023 = pd.read_excel('data/data_league_2023.xlsx')
df_league_2024 = pd.read_excel('data/data_league_2024.xlsx')
df_league_2024['MES'] = df_league_2024['DATA'].apply(lambda x: x.month)
df_league_2024['ANO'] = df_league_2024['DATA'].apply(lambda x: x.year)

#df_league = pd.concat([df_league_2023, df_league_2024])
```

```python
df_league_2024
```

```python
df_rodadas_no_mes = df_league_2024.groupby(['MES','RODADA']).agg({'RODADA':'count'}).rename({'RODADA':'PARTIDAS_JOGADAS'},axis=1).reset_index().drop('PARTIDAS_JOGADAS', axis=1).groupby('MES').agg({'RODADA':'count'}).rename({'RODADA':'TOTAL_RODADAS_MES'},axis=1).reset_index()
```

```python
df_lista_presenca = df_league_2024.groupby(['PLAYER', 'MES','RODADA']).agg({'RODADA':'count'}).rename({'RODADA':'PARTIDAS_JOGADAS'},axis=1).reset_index().drop('PARTIDAS_JOGADAS', axis=1)
```

```python
df_lista_presenca = df_lista_presenca.groupby(['PLAYER', 'MES']).agg({'RODADA':'count'}).rename({'RODADA':'PRESENCAS_NO_MES'},axis=1).reset_index()
```

```python
dff_lista_presenca = pd.merge(df_lista_presenca, df_rodadas_no_mes, how='left', on='MES')
dff_lista_presenca = dff_lista_presenca[['PLAYER', 'MES', 'TOTAL_RODADAS_MES', 'PRESENCAS_NO_MES']]
dff_lista_presenca['FALTAS'] = dff_lista_presenca['TOTAL_RODADAS_MES'] - dff_lista_presenca['PRESENCAS_NO_MES']
dff_lista_presenca['FALTAS_PERC'] = round((dff_lista_presenca['FALTAS'] / dff_lista_presenca['TOTAL_RODADAS_MES'])*100, 2)
```

```python
dff_lista_presenca.to_excel('lista_faltas.xlsx', index=False)
```
