```python
import pandas as pd
```

```python
df_copa_2023 = pd.read_excel('data/data_copa_2023.xlsx')
df_copa_2024 = pd.read_excel('data/data_copa_2024.xlsx')

df_copas = pd.concat([df_copa_2023, df_copa_2024])
```

```python
df_copas['ANO'] = df_copas['DATA'].apply(lambda x: x.year)
```

```python
df_copas_campeos = df_copas[df_copas['PTS']>=50]
```

```python
df_copas_campeos.groupby(['PLAYER']).agg({'PLAYER':'count'}).rename({'PLAYER':'TITULOS'},axis=1).reset_index().sort_values('TITULOS',ascending=False)
```
